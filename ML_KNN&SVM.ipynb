```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
ss = StandardScaler()
X_scaled = ss.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=45)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=7)
```

```python
KNN.fit(X_train,y_train)
knn_pred = KNN.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print("Accuracy of KNN model is",accuracy_score(y_test,knn_pred)*100)
```

```python
from sklearn.svm import SVC
linear = SVC(kernel='linear')
poly = SVC(kernel='poly')
rbf = SVC()
sigmoid = SVC(kernel='sigmoid')
precomputed = SVC(kernel='precomputed')
```

```python
linear.fit(X_train,y_train)
liner_pred=linear.predict(X_test)
print("Accuracy score of SVM model with linear kernel is",accuracy_score(y_test,liner_pred)*100,'%')
```

```python
poly.fit(X_train,y_train)
poly_pred = poly.predict(X_test)
print("Accuracy score of SVM model with poly kernel is",accuracy_score(y_test,poly_pred)*100,'%')
```

```python
rbf.fit(X_train,y_train)
rbf_pred = rbf.predict(X_test)
print("Accuracy score of SVM model with rbf kernel is",accuracy_score(y_test,rbf_pred)*100,'%')
```

```python
sigmoid.fit(X_train,y_train)
sig_pred = sigmoid.predict(X_test)
print("Accuracy score of SVM model with sigmoid kernel is",accuracy_score(y_test,sig_pred)*100,'%')
```
